```python
response = requests.get(url, headers=dict_header).content
```

```python
response
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup
```

```python
noticias = soup.find('', class_='icon-list loop-ul-latest loop-ul-in-body loop-ul-of-any')
```

```python
noticias
```

```python
rows = noticias.find_all('a')
```

```python
for i in rows:
  print(i)
```

```python
print(rows[1])
```

```python
df_news = pd.DataFrame()
tit = []
subt = []
for row in rows[1:]:
    for col in row.find_all('p'):
      subt.append(col.text)
      for col2 in row.find_all('h2'):
        tit.append(col2.text)
df_news['Ttulo'] = tit 
df_news['Subttulo'] = subt 
```
